## Description
> ##### The Hangman program randomly selects a secret word from a list of secret words.
> ##### The Game: Here, a random word (a fruit name) is picked up from our collection and the player gets limited chances to guess the fruit name correctly.
> ##### When a letter in that word is guessed correctly, that letter position in the word is made visible in the output.
> ##### In this way, all letters of the word are to be guessed before all the chances are over.
> ##### For convenience, the number of chances equals the word length + 2.
> ##### For example, word to be guessed is mango, then user gets 5 + 2 = 7 chances, as mango is a five letter word.
> ##### Make game repeats until player ends the game.
> ##### Bonus : Create a function that takes 3 parameters `Easy`, `Medium` and `Hard` where `Medium` has more difficult words to be guessed than `Easy` and so on.

## 1: Import 'random' and create the fruit list

## 2: Hide the letters of the secret word to prevent the player from seeing the secret word

```python
import random

fruits = ["apple", "mango", "grape", "lemon", "melon"]

random_word = random.choice(fruits)

word = []

for letter in random_word:
    word.append("_")

print(word)
```

## 3: Add the number of chances (The number of chances equals the word length + 2)

```python
import random

fruits = ["apple", "mango", "grape", "lemon", "melon"]

random_word = random.choice(fruits)

word = []

for letter in random_word:
    word.append("_")

chances = len(random_word) + 2

print("Word:", word)
print("Chances:", chances)
```

## 4: Ask the player to guess a letter

```python
import random

fruits = ["apple", "mango", "grape", "lemon", "melon"]

random_word = random.choice(fruits)

word = []

for letter in random_word:
    word.append("_")

chances = len(random_word) + 2

guess = input("Guess a letter: ").lower()

if guess in random_word:
    print("Correct guess")
else:
    print("Wrong guess")
    chances -= 1
```

## 5: show up correctly guessed letter

```python
import random

fruits = ["apple", "mango", "grape", "lemon", "melon"]

random_word = random.choice(fruits)

word = []

for letter in random_word:
    word.append("_")

chances = len(random_word) + 2

guess = input("Guess a letter: ").lower()

if guess in random_word:
    print("Correct guess")

    for i in range(len(random_word)):

        if random_word[i] == guess:
            word[i] = guess

else:
    print("Wrong guess")
    chances -= 1

print("Word:", " ".join(word))
print("Chances left:", chances)
```

## 6: Allow player to keep guessing while game isn't stopping after one guess

```python
import random

fruits = ["apple", "mango", "grape", "lemon", "melon"]

random_word = random.choice(fruits)

word = []

for letter in random_word:
    word.append("_")

chances = len(random_word) + 2

while chances > 0 and "_" in word:

    print("Word:", " ".join(word))
    print("Chances left:", chances)

    guess = input("Guess a letter: ").lower()

    if guess in random_word:

        print("Correct guess")

        for i in range(len(random_word)):

            if random_word[i] == guess:
                word[i] = guess

    else:

        print("Wrong guess")
        chances -= 1
```
